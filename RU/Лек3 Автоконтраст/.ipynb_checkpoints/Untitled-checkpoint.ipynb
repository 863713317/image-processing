```python
from skimage.io import imread, imsave, imshow
from skimage import img_as_ubyte
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = imread('parrot.png', as_grey=True)
img = img_as_ubyte(img)
img = img // 2 + 55
```

```python
plt.hist(img.ravel(), bins = 255, range=[0, 256]);
```

```python
imshow(img)
```

```python
img = img - img.min()
plt.hist(img.ravel(), bins = 255, range=[0, 256]);
```

```python
img.dtype
```

```python
img = img * np.uint8(255 / (img.max() - img.min()))
plt.hist(img.ravel(), bins = 255, range=[0, 256]);
```
