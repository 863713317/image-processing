```python
from skimage.io import imread, imshow
import numpy as np
from skimage.morphology import watershed
import matplotlib.pyplot as plt
from skimage import img_as_ubyte
%matplotlib inline
```

```python
img = img_as_ubyte(imread('coins.jpg', as_grey= True))
imshow(img)
```

```python
markers =  np.zeros(img.shape, dtype = np.uint8)
markers[150, 210] = 1
markers[50, 250] = 2
markers[100, 250] = 3
markers[120:170, 160:180] = 4
markers[200:250, 0:50] = 5
markers[250, 150] = 6
lbl_img = watershed(255 - img, markers = markers, mask = img > 20)
```

```python
plt.imshow(lbl_img, cmap=plt.cm.nipy_spectral)
```

```python
lbl_img.max()
```
