```python
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = imread('panther.jpeg')
print(img.shape)

plt.figure(figsize=(5,5))
imshow(img)
```

The peacock contains many colors

```python
plt.figure(figsize=(17,7))

plt.subplot(131),imshow(img[...,0])
plt.title('RED')

plt.subplot(132),imshow(img[...,1])
plt.title('GREEN')

plt.subplot(133),imshow(img[...,2])
plt.title('BLUE')
```

```python
imshow(img[...,0])
```

Peacock feathers consists of blue, green and other colors.
and the background consist of feathers with many colors.
so despite of the paeacock  form the backgrund is difficult.
But we see that the peacock in red and green channel is black and the background is gray . 
Now we can get the peacock mask for the red chanel.

```python
plt.figure(figsize=(5,5))
imshow(img[...,0] > 180)

plt.figure(figsize=(5,5))
imshow(np.bitwise_not(img[...,0] > 195))
```

Converted the image from RGB to gray 

```python
from skimage.color import rgb2gray
plt.figure(figsize=(5,5))
imshow(rgb2gray(img))
```

  Now transform the image from RGB to HSV

```python
from skimage.color import rgb2hsv
plt.figure(figsize=(15,6))
img_hsv = rgb2hsv(img)

plt.subplot(131),imshow(img_hsv[...,0])
plt.title('HUE')

plt.subplot(132),imshow(img_hsv[...,1])
plt.title('SATURATION')

plt.subplot(133),imshow(img_hsv[...,2])
plt.title('VALUE')
```

   We use And  mask

```python
plt.figure(figsize=(5,5))
gre_color = np.bitwise_and(img_hsv[...,0] > 0.4, img_hsv[...,0] < 0.7)
imshow(gre_color)
plt.title('Gre_color')
```

```python
from skimage.morphology import opening
a = np.ones((4,4))
imshow(opening(gre_color, a))
plt.title('Gre_color')
```

```python
plt.hist(img_hsv[...,1].ravel(), bins=100);
```

   We use OR mask

```python
plt.figure(figsize=(5,5))
gree_color = img_hsv[...,1] > 0.9
imshow(gree_color)
plt.title('Gree_color')
```

     Using mask again

```python
plt.figure(figsize=(5,5))
mask = np.bitwise_not(gre_color, gree_color)
imshow(mask)
plt.title('All together')
```

```python
from scipy.ndimage.morphology import binary_fill_holes
```

```python
imshow(binary_fill_holes(mask))
```
