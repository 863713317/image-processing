```python
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = imread('yellow_fish.jpg')
print(img.shape)
plt.figure(figsize=(5,5))
imshow(img)
```

```python
plt.figure(figsize=(15,6))
plt.subplot(131),imshow(img[...,0])
plt.title('RED')

plt.subplot(132),imshow(img[...,1])
plt.title('GREEN')

plt.subplot(133),imshow(img[...,2])
plt.title('BLUE')
```

The fish is yellow and the background consists of black and white.
we see that the fish in blue channel is black but the background is black,white and gray.
Now we can get the parrot mask.

```python
plt.figure(figsize=(5,5))
a=img[...,2] >55
imshow(a)
#imshow(img[...,2] >55)   # threshold 

plt.figure(figsize=(5,5))
imshow(np.bitwise_not(img[...,2] > 65)) #  mask with not 
```

```python
from skimage.color import rgb2gray
plt.figure(figsize=(5,5))
imshow(rgb2gray(img))
```

```python
from skimage.color import rgb2hsv
plt.figure(figsize=(15,6))
img_hsv = rgb2hsv(img)

plt.subplot(131),imshow(img_hsv[...,0])
plt.title('HUE')

plt.subplot(132),imshow(img_hsv[...,1])
plt.title('SATURATION')

plt.subplot(133),imshow(img_hsv[...,2])
plt.title('VALUE')
```

Now we use the and mask for img_hsv[...,0]

```python
plt.figure(figsize=(5,5))
hue_yellow_color = np.bitwise_and(img_hsv[...,0] > 0.05, img_hsv[...,0] < 0.16)
imshow(hue_yellow_color)
plt.title('hue_yellow_color')
```

```python
plt.figure(figsize=(5,5))
saturation_fish = img_hsv[...,1] > 0.60
imshow(saturation_fish)
plt.title('sat_yellow_color')
```

```python
plt.figure(figsize=(5,5))
hue_yellow_color = np.bitwise_and(saturation_fish, hue_yellow_color)
imshow(hue_yellow_color)
plt.title('hue_yellow_color')
```
