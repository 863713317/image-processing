```python
from skimage.io import imread, imshow
from skimage.filters import gaussian
from numpy.fft import fft2, ifft2, fftshift, ifftshift
import numpy as np
import skimage
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize = (15,10))
im = imread('rodina.jpg', as_grey=True)
# im.astype(np.float)
```

```python
plt.figure(figsize = (10,10))
sigma = 1.5
IM = fft2(im)
filt_gauss = np.zeros(im.shape)
filt_gauss[im.shape[0] // 2, im.shape[1] // 2] = 1
filt_gauss = gaussian(filt_gauss, sigma = sigma)
FILT = fft2(filt_gauss)
# im.astype(np.float)
plt.subplot(121)
imshow(im, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.subplot(122)
imshow(np.log(1 + abs(fftshift(fft2(im)))), cmap=plt.get_cmap('gray'), vmin=0, vmax=10)
```

```python
plt.figure(figsize = (10,10))
plt.subplot(121)
imshow(ifft2(IM * abs(FILT)).real, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.subplot(122)
imshow(np.log(1 + abs(fftshift(IM * FILT))), cmap=plt.get_cmap('gray'), vmin=0, vmax=10)
```
